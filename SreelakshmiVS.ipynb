```python
#a
import matplotlib.pyplot as plt
df = pd.read_csv('diabetes_model.csv')
glucose_age_subset = df.loc[45:60, ['Glucose', 'Age']]
print("\nGlucose level and Age for samples 45 to 60:")
glucose_age_subset
```

```python
#b
bp_count = df[df['BloodPressure'] > 120].shape[0]
print("\nNumber of patients having blood pressure above 120:",bp_count)
```

```python
#c
min = df['Insulin'].min()
max = df['Insulin'].max()
print("Minimum Insulin level:", min)
print("Maximum Insulin level:", max)
```

```python
#d
plt.figure(figsize=(7, 7))
plt.hist(df['Age'], bins=20, color='red', edgecolor='black')
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#e
plt.figure(figsize=(7, 7))
plt.scatter(df['BMI'], df['Glucose'], alpha=0.5, color='pink')
plt.title('Scatter Plot of BMI vs Glucose')
plt.xlabel('BMI')
plt.ylabel('Glucose')
plt.show()
```
